1. Group the data by the `Category` column and calculate the following aggregate statistics for each category:
   - Total quantity sold.
   - Average price per unit.
   - Maximum quantity sold in a single transaction.
2. Identify the top-selling product in each category based on the total quantity sold.
3. Find the date on which the highest total sales (quantity * price) occurred.

```python
(sales_data
 .groupby('Category')
 .agg({'Quantity':['sum'],
       'Price':['mean'],
       'Quantity':['max']})).reset_index()
```

```python
top_selling_product=(sales_data
 .groupby(['Category', 'Product'])['Quantity'].sum()).reset_index()

top_selling_product.loc[top_selling_product.groupby('Category')['Quantity'].idxmax()]
```

```python
sales_data['Total_Sales']=sales_data.Price*sales_data.Quantity

daily_sales=sales_data.groupby('Date')['Total_Sales'].sum().reset_index()

daily_sales.loc[daily_sales['Total_Sales'].idxmax()]
```


1. Group the data by `CustomerID` and filter out customers who have made less than 20 orders.
2. Identify customers who have ordered products with an average price per unit greater than $120.
3. Find the total quantity and total price for each product ordered, and filter out products that have a total quantity less than 5 units.

```python
customer_orders=pd.read_csv('C:\\Users\\User\\OneDrive\\Рабочий стол\\Python Hometasks\\customer_orders.csv')
customer_orders
```

```python
total=customer_orders.groupby('CustomerID')['OrderID'].count().reset_index()

result=total[total['OrderID']<20]
result
```

```python
average_price=customer_orders.groupby('Product')['Price'].mean().reset_index()

result=average_price[average_price['Price']>120].reset_index()
result
```

```python
total=(customer_orders
 .groupby('Product')
 .agg(Total_quantity=('Quantity','sum'),
      Total_price=('Price', 'sum'))).reset_index()

result=total[total['Total_quantity']<5].reset_index()
result
```
