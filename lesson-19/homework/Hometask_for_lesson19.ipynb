```python

survived=sns.barplot(x='sex', y='survived', data=titanic, hue='class', palette='coolwarm')
```

```python
import matplotlib.pyplot as plt

sns.barplot(x='who', y='fare', data=titanic, hue='survived', palette='coolwarm')
```
