```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://quotes.toscrape.com/page/{}/"
page = 1
all_quotes = []

while True:
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    quotes = soup.find_all("div", class_="quote")
    if not quotes:
        break  # Sahifa tugagan

    for quote in quotes:
        text = quote.find("span", class_="text").get_text(strip=True)
        author = quote.find("small", class_="author").get_text(strip=True)
        tags = [tag.get_text(strip=True) for tag in quote.find_all("a", class_="tag")]
        all_quotes.append([text, author, ", ".join(tags)])

    page += 1

# CSV faylga yozish
with open("quotes.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Iqtibos", "Muallif", "Teglar"])
    writer.writerows(all_quotes)

print("✅ 'quotes.csv' fayl yaratildi.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

BASE_URL = "https://books.toscrape.com/catalogue/page-{}.html"
books = []

def get_star_rating(class_list):
    for c in class_list:
        if c in ["One", "Two", "Three", "Four", "Five"]:
            return c
    return "No Rating"

for page in range(1, 51):  # 50 sahifa bor
    print(f"Scraping page {page}...")
    url = BASE_URL.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    articles = soup.select("article.product_pod")
    for article in articles:
        title = article.h3.a['title']
        price = article.select_one(".price_color").text.strip()
        rating_class = article.p['class']
        rating = get_star_rating(rating_class)

        books.append({
            "Title": title,
            "Price": price,
            "Rating": rating
        })

# Natijani CSV faylga yozamiz
with open("books.csv", mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["Title", "Price", "Rating"])
    writer.writeheader()
    writer.writerows(books)

print("✅ Ma'lumotlar 'books.csv' fayliga saqlandi.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Amazon qidiruv URL
def get_amazon_data(search_term):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9"
    }
    
    # Qidiruv so'zini URLga qo'shish
    url = f"https://www.amazon.com/s?k={search_term.replace(' ', '+')}"
    
    # HTTP so'rov yuborish
    response = requests.get(url, headers=headers)
    
    # Agar so'rov muvaffaqiyatli bo'lsa
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'lxml')
        results = []
        
        # Mahsulotlar ro'yxatini topish
        listings = soup.find_all('div', {'data-component-type': 's-search-result'})
        
        for listing in listings:
            item = {}
            
            # Mahsulot nomi
            name = listing.find('span', class_='a-size-medium a-color-base a-text-normal')
            item['Name'] = name.text.strip() if name else 'N/A'
            
            # Narx
            price = listing.find('span', class_='a-offscreen')
            item['Price'] = price.text.strip() if price else 'N/A'
            
            # Reyting
            rating = listing.find('span', class_='a-icon-alt')
            item['Rating'] = rating.text.split()[0] if rating else 'N/A'
            
            # Sharhlar soni
            reviews = listing.find('span', class_='a-size-base')
            item['Reviews'] = reviews.text.strip() if reviews else 'N/A'
            
            # Mahsulot havolasi
            link = listing.find('a', class_='a-link-normal s-no-outline')
            item['URL'] = 'https://www.amazon.com' + link['href'] if link else 'N/A'
            
            results.append(item)
        
        # Natijalarni DataFrame'ga aylantirish
        df = pd.DataFrame(results)
        df.to_csv('amazon_products.csv', index=False)
        print("Ma'lumotlar 'amazon_products.csv' fayliga saqlandi!")
        return df
    else:
        print(f"Xatolik: {response.status_code}")
        return None

# Misol uchun qidiruv
search_term = "laptop"
get_amazon_data(search_term)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

def scrape_linkedin_jobs(search_term="Python developer", max_results=10, scroll_count=3):
    # Selenium sozlamalari
    options = Options()
    options.add_argument("--headless")  # Brauzerni ko'rinmas holatda ishlatish
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    
    driver = webdriver.Chrome(options=options)
    
    # LinkedIn Jobs URL
    url = f"https://www.linkedin.com/jobs/search/?keywords={search_term.replace(' ', '%20')}"
    driver.get(url)
    
    # Ma'lumotlarni saqlash uchun ro'yxat
    jobs = []
    
    try:
        # Sahifadagi ish e'lonlarini yuklanishini kutish
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "jobs-search__results-list"))
        )
        
        # Sahifani bir necha marta skrol qilish
        for _ in range(scroll_count):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # Yangi e'lonlar yuklanishini kutish
        
        # Ish e'lonlari ro'yxatini topish
        job_listings = driver.find_elements(By.CLASS_NAME, "base-card")
        
        for job in job_listings[:max_results]:
            try:
                # Ish nomi
                title = job.find_element(By.CLASS_NAME, "base-search-card__title").text.strip()
                
                # Kompaniya nomi
                company = job.find_element(By.CLASS_NAME, "base-search-card__subtitle").text.strip()
                
                # Joylashuv
                location = job.find_element(By.CLASS_NAME, "job-search-card__location").text.strip()
                
                # Havola
                link = job.find_element(By.CLASS_NAME, "base-card__full-link").get_attribute("href")
                
                jobs.append({
                    "Title": title,
                    "Company": company,
                    "Location": location,
                    "URL": link
                })
            except Exception as e:
                print(f"Ish e'lonini o'qishda xato: {e}")
                continue
        
        # Natijalarni CSV fayliga saqlash
        df = pd.DataFrame(jobs)
        df.to_csv('linkedin_jobs.csv', index=False)
        print("Ma'lumotlar 'linkedin_jobs.csv' fayliga saqlandi!")
        
    except Exception as e:
        print(f"Sahifani yuklashda xato: {e}")
    
    finally:
        driver.quit()
    
    return jobs

# Misol uchun qidiruv
scrape_linkedin_jobs("Python developer", max_results=10, scroll_count=3)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Selenium settings
chrome_options = Options()
# Comment out headless mode for debugging
# chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920,1080")

try:
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
except Exception as e:
    print(f"Failed to initialize WebDriver: {e}")
    exit()

# Function to wait for page load
def wait_for_page_load(driver, timeout=10):
    WebDriverWait(driver, timeout).until(
        EC.presence_of_element_located((By.TAG_NAME, "body"))
    )

# Open Airbnb
try:
    driver.get("https://www.airbnb.com")
    wait_for_page_load(driver)
except Exception as e:
    print(f"Failed to load Airbnb: {e}")
    driver.quit()
    exit()

# Search for a city
search_query = "Tashkent, Uzbekistan"  # Include country for specificity
try:
    # Wait for and interact with the search box
    search_box = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "input[data-testid='structured-search-input-field-query']"))
    )
    search_box.click()
    search_box.send_keys(search_query)
    
    # Wait for and select the first suggestion (if any)
    try:
        suggestion = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "div[data-testid='structured-search-input-field-query-panel'] li"))
        )
        suggestion.click()
    except:
        print("No suggestions found, proceeding with search...")
    
    # Wait for and click the search button
    search_button = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "button[data-testid='structured-search-input-search-button']"))
    )
    search_button.click()
except Exception as e:
    print(f"Error during search: {e}")
    driver.quit()
    exit()

# Wait for search results to load
try:
    WebDriverWait(driver, 15).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div[data-testid='card-container']"))
    )
except Exception as e:
    print(f"Error loading search results: {e}")
    # Save page source for debugging
    with open("page_source.html", "w", encoding="utf-8") as f:
        f.write(driver.page_source)
    driver.quit()
    exit()

# Scrape listings
listings = driver.find_elements(By.CSS_SELECTOR, "div[data-testid='card-container']")

if not listings:
    print("No listings found.")
else:
    for listing in listings[:5]:  # Limit to first 5 listings
        try:
            # Extract title/location (updated selector)
            title = listing.find_element(By.CSS_SELECTOR, "div[data-testid='listing-card-title']").text if listing.find_elements(By.CSS_SELECTOR, "div[data-testid='listing-card-title']") else "Title not available"
            
            # Extract price (updated selector)
            price = listing.find_element(By.CSS_SELECTOR, "span[data-testid='price-availability-row'] span").text if listing.find_elements(By.CSS_SELECTOR, "span[data-testid='price-availability-row'] span") else "Price not available"
            
            # Extract rating (updated selector)
            rating = listing.find_element(By.CSS_SELECTOR, "span[data-testid='listing-card-rating']").text if listing.find_elements(By.CSS_SELECTOR, "span[data-testid='listing-card-rating']") else "No rating"

            print(f"Title/Location: {title}")
            print(f"Price: {price}")
            print(f"Rating: {rating}")
            print("-" * 50)
        except Exception as e:
            print(f"Error scraping listing: {e}")
            continue

# Clean up
driver.quit()
```
