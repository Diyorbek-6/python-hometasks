```python
import pandas as pd

data = [['NY', 'blue', 'Steak', 30, 165, 4.6],
        ['TX', 'green', 'Lamb', 2, 70, 8.3],
        ['FL', 'red', 'Mango', 12, 120, 9.0],
        ['AL', 'white', 'Apple', 4, 80, 3.3],
        ['AK', 'gray', 'Cheese', 32, 180, 1.8],
        ['TX', 'black', 'Melon', 33, 172, 9.5],
        ['TX', 'red', 'Beans', 69, 150, 2.2]]

columns = ['state', 'color', 'food', 'age', 'height', 'score']
index = ['Jane', 'Niko', 'Aaron', 'Penelope', 'Dean', 'Christina', 'Cornelia']

df = pd.DataFrame(data, columns=columns, index=index)
```

```python

def highlight_cols(s):
    return ['background-color: yellow;  color:black' if s.name in ['age', 'height', 'color'] else '' for i in s]

df.style.apply(highlight_cols, subset=['age', 'height', 'color'])
```

```python

def highlight_rows(s):
    if s.name in ['Aaron', 'Dean']: 
        return ['background-color: yellow; color: black' for i in s]
    return ['' for i in s]

styled_1 = df.style.apply(highlight_rows, axis=1)
styled_1
```

```python

def highlight_index(data):
    color_map = pd.DataFrame('', index=data.index, columns=data.columns)
    color_map.loc['Aaron', ['color', 'age', 'height' ]] = 'background-color: yellow; color: black'
    color_map.loc['Dean', ['color', 'age', 'height']] = 'background-color: yellow; color: black'
    return color_map

styled_2 = df.style.apply(highlight_index, axis=None)
styled_2
```
