```python
#ToDoList Application
class Taskclass:
    def __init__(self, title, description, due_date):
        self.title = title
        self.description = description
        self.due_date = due_date
        self.completed = False  

    def mark_completed(self):
        self.completed = True 

    def __str__(self):
        status = "âœ… COMPLETED" if self.completed else "âŒ NOT COMPLETED"
        return f"ğŸ¯ {self.title} ({self.due_date}) - {status}\n {self.description}"

class ToDoList:
    def __init__(self):
        self.task = []

    def add_task(self, task):
        self.task.append(task)
        print(f"âœ… Task '{task.title}' added successfully!")

    def mark_task_complete(self, title):
        for task in self.task:
            if task.title.lower() == title.lower():
                task.mark_completed()  
                print(f'âœ”ï¸ Task "{title}" marked as complete!')
                return
        print(f"âŒ Task '{title}' not found!")

    def list_tasks(self):
        if not self.task:
            print("âŒ No tasks available")
        else:  
            print("ğŸ“ All tasks:")
            for task in self.task:
                print(task)

    def incomplete_tasks(self):
        incomplete_tasks = [task for task in self.task if not task.completed]
        if not incomplete_tasks:
            print("ğŸ‰ No incomplete tasks")
        else:
            print("âŒ› Incomplete tasks:")
            for task in incomplete_tasks:
                print(task)

def main():
    todo_list = ToDoList()

    while True:
        print("\nTo-Do List Menu")
        print("1ï¸âƒ£ Add task")
        print("2ï¸âƒ£ Mark task as complete")
        print("3ï¸âƒ£ List all tasks")
        print("4ï¸âƒ£ List incomplete tasks")
        print("5ï¸âƒ£ Exit")

        choice = input("Choose an option: ")  

        if choice == "1":
            title = input("ğŸ“ Enter task title: ")
            description = input("ğŸ“„ Enter task description: ")
            due_date = input("ğŸ“† Enter due date (YYYY-MM-DD): ")
            task = Taskclass(title, description, due_date)  
            todo_list.add_task(task)

        elif choice == "2":
            title = input("Enter task title to mark as complete: ")
            todo_list.mark_task_complete(title)

        elif choice == "3":
            todo_list.list_tasks()

        elif choice == "4":
            todo_list.incomplete_tasks()

        elif choice == "5":
            print("ğŸ‘‹ Exiting... Goodbye!!!")
            break  

        else:
            print("âŒ Error. Please enter a number between 1 and 5!")

if __name__ == "__main__": 
    main()
```
